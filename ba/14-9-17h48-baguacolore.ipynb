```python
def f(x):
    if x<0:
        return sqrt(-x^2-2*x)
    else:
        return -sqrt(-x^2+2*x)
a=plot(f,(x,-2,2),color='white',fill=sqrt(4-x^2),
fillcolor="black",alpha=0.1)
b=plot(f,(x,-2,2),color="white",fill=-sqrt(4-x^2),
fillcolor="red",alpha=0.1)
show(a+b+pp+pc,figsize=4,fontsize=8,
axes=True,frame=True)
```

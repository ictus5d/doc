###### Quadrature du cercle

Elle peut s'obtenir de deux façons, La version Euler et ma façon que j'ai découverte en la faisant.

```python
E = text("Euler : $\\frac{\\sqrt{\\pi}}{2}$", (-0.01,0), 
         color="green",fontsize=18)
p = point((0.705,0.705), color="red",pointsize=10, zorder=5)
cp=line([(-0.886,-0.886),(-0.886,0.886),(0.886,0.886)])
cq=line([(0.886,0.886),(0.886,-0.886),(-0.886,-0.886)])
ce = circle((0,0),1,color="green",zorder=-5,thickness=1)
show(ce+cp+cq+E+p,xmin=-1,xmax=1,ymin=-1,ymax=1,gridlines="minor",
gridlinesstyle=dict(color="orange",linestyle="-",alpha=0.2),
axes=False,frame=True,figsize=5)
```

```python
G = text("Inverse : $\\frac{2}{\\sqrt{\\pi}}$", (-0,0),
         color="red",fontsize=15)
p = point((0.8,0.8), color="black",pointsize=10, zorder=5)
cp=line([(-1,-1),(-1,1),(1,1)])
cq=line([(1,1),(1,-1),(-1,-1)]);r=2/sqrt(n(pi))
ci = circle((0,0),r,color="red",zorder=-5,thickness=1)
show(ci+cp+cq+G+ce+p,xmin=-1.2,xmax=1.2,ymin=-1.2,ymax=1.2,gridlines="minor",
gridlinesstyle=dict(color="orange",linestyle="-",alpha=0.2),
axes=False,frame=True,figsize=5)
```

Il est plus facile de tracer un carré de 1 et ensuite un cercle qui sera un point précis de la diagonale, 5 unités, cercle sur la quatrième unité qui est la quadrature parfaite, ce qui démontre que Lindemann a enfumé les quadrateurs !

Dans le cas des deux cercles les points sur la diagonales sont à la même quatrième unité : Les équations peuvent être échangées et donneront le même résultat en unité, le cercle d'Euler étant plus petit et moins précis.

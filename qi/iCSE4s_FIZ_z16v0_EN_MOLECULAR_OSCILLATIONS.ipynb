```python
var('A,w,t,f,x,a,v,F,m')
#amplitude of the oscillation
A = 1.12*(10^-11) 
#angular frequency
w = 1.05*(10^14) 
#initial phase
f = 0.087 
#displacement-time equation
x(t) = A*sin(w*t+f)
#velocity-time equation
v(t) = x.diff(t)
#acceleration-time equation
a(t) = v.diff(t)
#force 
F(t) = m*a(t)

m = 1.0
    
wykres1=plot(m*a(t),(t,0,1e-13),figsize=(3,2),color="red",thickness=1.0, fill=True, fillcolor="gold",legend_label=('Force acts on a molecule)'))

wykres2=plot(x(t),(t,0,1e-13),figsize=(3,2),color="red",thickness=1.0,linestyle="--", legend_label=('x(t)'))
wykres3=plot(v(t),(t,0,1e-13),figsize=(3,2),color="green",thickness=1.0, linestyle="-.", legend_label=('v(t)'))
wykres4=plot(a(t),(t,0,1e-13),figsize=(3,2),color="blue",thickness=1.0, legend_label=('a(t)'))
html.table([["siła","$x(t)$","$v(t)$","$a(t)$"],[ wykres1,wykres2,wykres3,wykres4]])
```
